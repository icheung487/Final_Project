```python
import pandas as pd
import glob
from datetime import datetime,date
import numpy as np
```

```python
path = "/Users/Anina/Desktop/COVID-19/test_sample"
```

```python
all_files = glob.glob(path + "/*.csv")
```

```python
all_files
```

```python
li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)
```

```python
df = pd.concat(li, axis=0, ignore_index=True)
```

```python
df
```

```python
#For Loop to filter the columns and replace the null values
columns_list = [['Incidence_Rate','Incident_Rate'],
                ['Case-Fatality_Ratio','Case_Fatality_Ratio']]

for o,n in columns_list:
    df[f'clean_{o}'] = np.where(df[o].isna(),df[n],df[o])
    
```

```python
df
```

```python
#Remove Time Stampe from column 'last_update'
df['Last_Update'] = pd.to_datetime(df['Last_Update']).dt.normalize()
df
```

```python
#created a new Df with new columns
new_df = df[['Admin2','Province_State','Country_Region',
            'Last_Update','Lat','Long_','Confirmed','Deaths',
            'Recovered','Active','clean_Incidence_Rate',
            'clean_Case-Fatality_Ratio']].copy()
new_df.tail()
```

```python
#Renaming Columns
#df.rename(columns={'oldName1': 'newName1', 'oldName2': 'newName2'}, inplace=True)
new_df.rename(columns={'clean_Incidence_Rate': 'Incident_Rate',
                      'clean_Case-Fatality_Ratio':'Case_Fatality_Ratio',
                      'Admin2': 'City',
                      'Confirmed':'Confirmed_cases'}, inplace= True)
new_df.tail()
```
