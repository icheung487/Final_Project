```python
all_files = glob.glob(path + "/*.csv")
all_files
```

```python
li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)
df = pd.concat(li, axis=0, ignore_index=True)
df.head()
```

```python
df.loc[df['Country_Region']=='US']
```

```python
#For Loop to filter the columns and replace the null values
columns_list = [['Province_State','Province/State'],
                 ['Country_Region','Country/Region'],
                 ['Last_Update','Last Update'],
                 ['Incident_Rate','Incidence_Rate'],
                ['Case_Fatality_Ratio','Case-Fatality_Ratio'],
                 ['Lat','Latitude'],
                 ['Long_','Longitude']]

for o,n in columns_list:
    df[f'clean_{o}'] = np.where(df[o].isna(),df[n],df[o])
    
```

```python
#created a new Df with new columns
new_df = df[['Admin2','clean_Province_State','clean_Country_Region',
            'clean_Last_Update','clean_Lat','clean_Long_','Confirmed',
            'Active']].copy()
new_df.head()
```

```python
#Renaming Columns
new_df.rename(columns={'Admin2': 'City',
                       'clean_Province_State': 'Province_State',
                       'clean_Country_Region': 'Country_Region',
                       'clean_Last_Update': 'Last_Update',
                       'Active': 'Cases_Per_Day',
                       'clean_Lat':'Lat',
                       'clean_Long_':'Long',
                      'Confirmed':'Confirmed_Cases'}, inplace= True)
new_df['Last_Update'] = pd.to_datetime(df['Last_Update']).dt.normalize()
new_df
```

```python
new_df.loc[new_df['Country_Region']=='China']
```

```python
country = pd.read_csv('people_fully_vaccinated.csv')
country
```

```python
country.rename(columns={'Day':'Last_Update1',
                       'Entity': 'Country_Region'}, inplace= True)
# country.iloc[0,2]
country
```

```python
country['Last_Update1'] = pd.to_datetime(country['Last_Update1'], errors='coerce')
country['Last_Update'] = country['Last_Update1']
country.iloc[0,-1]
```

```python
country.loc[country['Country_Region']=='China']
```

```python
merged_df = new_df.merge(country, on=['Country_Region','Last_Update'])
merged_df.head()
```

```python
merged_df.loc[merged_df['Country_Region']=='China']
```

```python
#created a new Df with new columns
final_df = merged_df[['City','Province_State',
               'Country_Region','Lat','Long',
               'Confirmed_Cases','Cases_Per_Day',
               'Last_Update','people_fully_vaccinated']].copy()
final_df.dropna(subset = ["Confirmed_Cases"], inplace=True)
final_df
```
