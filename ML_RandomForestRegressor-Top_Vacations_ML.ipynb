```python
#(DATAFRAME TO BE MERGED WITH PREDICTATED CONFIRMED CASES)
Final_df = Country_df.loc[Country_df['cases_per_day'] >=1]
                          
Final_df.head()
```

```python
Final_Country_df = Final_df.drop(['coordinates'], axis=1)
Final_Country_df
```

```python
len(Final_df)
```

```python
#Encoded the categorical data
encoded_df = pd.get_dummies(Final_Country_df, columns=["city","province_state","country_region","last_update",])
encoded_df.head()
```

```python
len(encoded_df) 
```

```python
# Define features set
X = encoded_df.copy()
X = encoded_df.drop('confirmed_cases', axis=1)
X.head
```

```python
X.shape
```

```python
# Define the target set.
y = encoded_df["confirmed_cases"]
y[:5]
```

```python
# # Splitting into Train and Test sets.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# # Fitting Random Forest Regression to the dataset
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 20, random_state = 42)
regressor = regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
y_pred
```

```python
accuracy = regressor.score(X_train,y_train)
accuracy
```

```python
print( 'Linear Regression Accuracy: ', accuracy*100,'%')
```

```python
print("Training Accuracy = ", regressor.score(X_train, y_train))
print("Test Accuracy = ", regressor.score(X_test, y_test))
```
