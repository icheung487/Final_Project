```python


# -----------------------------------------------------------------------------------
dates=[datetime.datetime(2020,1,22)+datetime.timedelta(dval) for dval in range(0,10)];
# -----------------------------------------------------------------------------------
print('starting')
#os.chdir(folder_path_download)
# Download daily added CSV files from https://github.com/CSSEGISandData/COVID-19/tree/master/csse_covid_19_data/csse_covid_19_daily_reports
for dateval in dates:
    url_string = 'https://github.com/CSSEGISandData/COVID-19/blob/master/csse_covid_19_data/csse_covid_19_daily_reports/'+ dateval.strftime('%m-%d-%Y')+'.csv?raw=true'
    r= requests.get(url_string)
    df=pd.read_csv(url_string,index_col=0)
    df.to_csv(dateval.strftime('%m-%d-%Y')+".csv")
    display(df.head())

print('ending')

```
